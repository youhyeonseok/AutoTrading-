```python
import pandas as pd
```

```python
apple_df = pd.read_csv('test.csv', index_col=0, parse_dates=True)
apple_df.head()
```

```python
import mplfinance as fplt

print("MPLFinance Version : {}".format(fplt.__version__))
```

```python
fplt.plot(
            apple_df,
            type='candle',
            title='Apple, March - 2020',
            ylabel='Price ($)'
        )
```

```python
fig = fplt.plot(
            apple_df.iloc[:100],
            type='candle',
            style='charles',
            title='Apple, March - 2020',
            ylabel='Price ($)',
            volume=True,
            ylabel_lower='Shares\nTraded',
            show_nontrading=True,
            returnfig = True
            )
```
